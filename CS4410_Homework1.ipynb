###Importing Modules

###Downloading CSV file from GitHub

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/Gingercapo/Diamond_dataset/main/diamonds.csv'
df = pd.read_csv(url)
```

```python
df.describe()
```

###Using the series method describe to calculate the descriptive statistics for the categorical data (cut, color, and clarity)

```python
df[['cut', 'color', 'clarity']].describe()
```

```python
for col in ['cut', 'color', 'clarity']:
    print(f"Unique values in {col}: {df[col].unique()}")
```
