##Loading matplotlib and diabetes dataset

```python
from sklearn import datasets
```

```python
%matplotlib inline
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

##Displaying the Dataset's Description

```python
print(diabetes.DESCR)
```

##Exploring the Data with Pandas

```python
import pandas as pd
```

```python
pd.set_option('display.precision', 4)
```

```python
diabetes_df = pd.DataFrame(diabetes.data,
                             columns=diabetes.feature_names)
```

```python
diabetes_df['Target'] = pd.Series(diabetes.target)
```

```python
diabetes_df.head() #Looking at the first five rows
```

```python
diabetes_df.describe() #Showing summary statistics
```

```python
sample_df = diabetes_df.sample(frac=0.1, random_state=17) #Taking a random sample of 10% of the data
```

```python
import matplotlib.pyplot as plt #Importing matplotlib
```

```python
import seaborn as sns #Importing seaborn
```

```python
sns.set_style('whitegrid') #Setting the style of the plots
```

```python
for feature in diabetes.feature_names:
    plt.figure(figsize=(8, 4.5))
    sns.scatterplot(data=sample_df, x=feature, y='Target', hue='Target',
                    palette='cool', legend=False) #Plotting the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    diabetes.data, diabetes.target, random_state=11) #Splitting the data into training and testing sets and using random state to reproduce results
```

```python
X_train.shape
```

```python
X_test.shape
```

##Training the Model

```python
from sklearn.linear_model import LinearRegression #Importing the Linear Regression model
```

```python
linear_regression = LinearRegression() #Creating an instance of the model
```

```python
linear_regression.fit(X=X_train, y=y_train)
```

```python
for i, name in enumerate(diabetes.feature_names):
    print(f'{name:>10}: {linear_regression.coef_[i]}') #Printing the coefficients for each feature
```

```python
linear_regression.intercept_ #Printing the intercept
```

##Testing the model

```python
predicted = linear_regression.predict(X_test) #Predicting the target values for the test set
```

```python
expected = y_test #Getting the expected values for the test set
```

```python
predicted[:5] #Printing the first five predicted values
```

```python
expected[:5] #Printing the first five expected values
```

##Visualizing the Expected vs. Predicted Values

```python
df = pd.DataFrame() #Creating a new dataframe
```

```python
df['Expected'] = pd.Series(expected) #Adding the expected values to the dataframe
```

```python
df['Predicted'] = pd.Series(predicted) #Adding the predicted values to the dataframe
```

```python
figure = plt.figure(figsize=(9, 9))

axes = sns.scatterplot(data=df, x='Expected', y='Predicted',
    hue='Predicted', palette='cool', legend=False)

start = min(expected.min(), predicted.min())

end = max(expected.max(), predicted.max())

axes.set_xlim(start, end)

axes.set_ylim(start, end)

line = plt.plot([start, end], [start, end], 'k--') #Plotting the line of best fit
```

##Regression Model Metrics

```python
from sklearn import metrics #Importing metrics
```

```python
metrics.r2_score(expected, predicted) #Printing the R^2 score
```

##Choosing the best Model

```python
from sklearn.linear_model import ElasticNet, Lasso, Ridge
```

```python
estimators = {
    'LinearRegression': linear_regression,
    'ElasticNet': ElasticNet(),
    'Lasso': Lasso(),
    'Ridge': Ridge()
} #Creating a dictionary of models
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
for estimator_name, estimator_object in estimators.items():
    kfold = KFold(n_splits=10, random_state=11, shuffle=True)
    scores = cross_val_score(estimator=estimator_object,
        X=diabetes.data, y=diabetes.target, cv=kfold,
        scoring='r2')
    print(f'{estimator_name:>16}: ' +
          f'mean of r2 scores={scores.mean():.3f}') #Printing the R^2 scores for each model
```

##Looking for a R^2 score as close to one as possible, Linear Regression is the closest
